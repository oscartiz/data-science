# Shoe Sales Dataset Exploration

This notebook provides a preliminary look at the `shoes_sales_dataset.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plotting style
sns.set_theme(style="whitegrid")

# Load the dataset
df = pd.read_csv('data/shoes_sales_dataset.csv')

# Display the first few rows
df.head()
```

```python
# Basic information about the dataset
df.info()
```

```python
# Descriptive statistics
df.describe()
```

## Visualizations

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Brand', order=df['Brand'].value_counts().index, palette='viridis')
plt.title('Number of Sales by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Brand', y='Revenue_USD', estimator=sum, errorbar=None, order=df.groupby('Brand')['Revenue_USD'].sum().sort_values(ascending=False).index, palette='mako')
plt.title('Total Revenue by Brand')
plt.xlabel('Brand')
plt.ylabel('Total Revenue (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Shoe_Type', order=df['Shoe_Type'].value_counts().index, palette='rocket')
plt.title('Number of Sales by Shoe Type')
plt.xlabel('Shoe Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Sales_Channel', y='Price_USD', palette='coolwarm')
plt.title('Price Distribution by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Price (USD)')
plt.show()
```
